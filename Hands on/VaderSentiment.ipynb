## Vader Sentiment
VADER (Valence Aware Dictionary and sEntiment Reasoner) is a lexicon and rule-based sentiment analysis tool that is specifically attuned to sentiments expressed in social media.

- Sentimental analysis can be performed using the polarity_scores() of the object generated by Sentimental Intensity Analyzer().
- The score of compound is the sensitivity index which is a combination of neg, neu, pos. (neg: Negative, neu: Neutral, pos:Positive)
- A compound has a value between -1 ~ 1 and is considered positive if it is 0.1 or higher
